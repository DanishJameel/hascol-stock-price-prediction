# LSTM Model ( Long Short Term Memory )


```python
# Split into training and testing sets
train_size = int(len(df) * 0.8)
test_size = len(df) - train_size
train, test = df[0:train_size,:], df[train_size:len(df),:]
```

```python

# Convert data into time-series format
def create_time_steps(length):
    return list(range(-length, 0))

def create_window(length):
    return [np.zeros(length) for i in range(df.shape[1])]

```

```python
def to_time_series_format(data, length):
    if len(data) <= length:
        raise ValueError("Input data must have at least `length + 1` rows.")

    time_steps = len(data) - length
    windows = [data.iloc[i:i + length].to_numpy() for i in range(time_steps)]
    time_series = np.array([np.append(w, d) for w, d in zip(windows, data.iloc[length:].to_numpy())])
    return time_series, time_steps
```

```python
def to_time_series_format(data, length):
    time_steps = len(data) - length
    windows = [data[i:i + length] for i in range(time_steps)]
    time_series = np.array([np.append(w, d) for w, d in zip(windows, data[length:])])
    return time_series, time_steps
```

```python
# Convert training and testing sets into time-series format
length = 10
train_time_series, train_time_steps = to_time_series_format(train, length)
test_time_series, test_time_steps = to_time_series_format(test, length)
```

```python

# Split into input and target sets
x_train, y_train = train_time_series[:, :-1], train_time_series[:, -1]
x_test, y_test = test_time_series[:, :-1], test_time_series[:, -1]
```

```python

# Reshape data into 3D format
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
# Define model
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(50, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(tf.keras.layers.LSTM(50))
model.add(tf.keras.layers.Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Train model
history = model.fit(x_train, y_train, epochs=20, batch_size=1)
```

```python
import matplotlib.pyplot as plt

# Predict the price for next 60 days
predicted_price = model.predict(x_test)

# Plot the actual and predicted prices
plt.plot(y_test, label='Actual Price')
plt.plot(predicted_price, label='Predicted Price')
plt.legend()
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
# Split the data into training and testing sets
train_data = data[:int(0.8*len(data))]
test_data = data[int(0.8*len(data)):]
```

```python
# Define the input and output sequence lengths
input_seq_len = 90
output_seq_len = 90
```

```python
# Convert the data into sequences
def create_sequence(data, input_seq_len, output_seq_len):
    X, y = [], []
    for i in range(len(data) - input_seq_len - output_seq_len + 1):
        X.append(data[i:i + input_seq_len])
        y.append(data[i + input_seq_len:i + input_seq_len + output_seq_len])
    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    return X, y
```

```python
train_X, train_y = create_sequence(train_data, input_seq_len, output_seq_len)
test_X, test_y = create_sequence(test_data, input_seq_len, output_seq_len)
```

```python
# Build the CNN model
model = Sequential()
model.add(Conv1D(32, 3, activation='relu', input_shape=(input_seq_len, 1)))
model.add(MaxPooling1D(2))
model.add(Flatten())
model.add(Dense(output_seq_len))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python

# Train the model
model.fit(train_X, train_y, epochs=100, batch_size=32, verbose=0)
```

```python
# Evaluate the model on the test data
score = model.evaluate(test_X, test_y, verbose=0)
print("Test MSE:", score)
```

# Autoencoder Deep learning Model  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Model
from keras.layers import Input, Dense
from sklearn.preprocessing import MinMaxScaler

# Load the stock price data
df = pd.read_csv("HASC Historical Data.csv")
# Drop the Irrelevant Columns 
df.drop(columns=["Change %", "Vol.","Open", "High", "Low"], inplace=True, axis=1)


df_price = df.Price
df = df.set_index('Date') # set date as index
scaler = MinMaxScaler() # initialize scaler
df = scaler.fit_transform(df) # scale data
df_scaled = df

# Split the data into training and test sets
train = df_scaled[:int(df_scaled.shape[0]*0.8), :]
test = df_scaled[int(df_scaled.shape[0]*0.8):, :]

# Define the input layer
input_layer = Input(shape=(train.shape[1],))

# Define the encoding layer
encoded = Dense(100, activation='relu')(input_layer)

# Define the decoding layer
decoded = Dense(train.shape[1], activation='sigmoid')(encoded)

# Define the autoencoder model
autoencoder = Model(input_layer, decoded)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model to the training data
autoencoder.fit(train, train, epochs=100, batch_size=32, shuffle=False)

# Use the trained autoencoder to make predictions on the test data
predictions = autoencoder.predict(test)

# Plot the actual and predicted stock prices
plt.plot(scaler.inverse_transform(test), label='Actual')
plt.plot(scaler.inverse_transform(predictions), label='Predicted')
plt.legend()
plt.show()
```

```python
test =  test
predictions = predictions

import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluation_metrics(test, predictions):
    mae = mean_absolute_error(test, predictions)
    mse = mean_squared_error(test, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(test, predictions)
    mape = np.mean(np.abs((test - predictions) / test)) * 100
    smape = np.mean(np.abs((test - predictions) / (np.abs(test) + np.abs(predictions))) * 200)
    
    return mae, mse, rmse, r2, mape, smape

if len(test) != len(predictions):
    min_len = min(len(test), len(predictions))
    test = test[:min_len]
    predictions = predictions[:min_len]

mae, mse, rmse, r2, mape, smape = evaluation_metrics(test, predictions)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot each metric as a bar
ax.bar(['MAE', 'MSE', 'RMSE', 'R2', 'MAPE', 'SMAPE'], [mae, mse, rmse, r2, mape, smape])

# Add labels and title
ax.set_xlabel('Metric')
ax.set_ylabel('Value')
ax.set_title('Evaluation Metrics')

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
# Convert Date column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as the index
df.set_index('Date', inplace=True)
```

```python

# Fit ARIMA model
model = ARIMA(df['Price'], order=(2,1,2))
model_fit = model.fit()

```

```python
# Print summary of ARIMA model
print(model_fit.summary())

# Predict future values
predictions = model_fit.predict(start=len(df), end=len(df)+30, typ='levels')
print(predictions)
```

```python
import matplotlib.pyplot as plt

# Plot actual vs predicted prices
plt.plot(df['Price'], label='Actual Price')
plt.plot(predictions, label='Predicted Price')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Actual vs Predicted Price')
plt.show()
```
